```python
from graphviz import Graph

g = Graph('G')
g.attr(rankdir='TB', size='20', splines='true', ranksep='0.5')
g.attr('node', shape='circle', rankdir='TB', style='bold', fontname = "verdana")

g.node('0', style='filled,bold', fixedsize='true', width='.9')
g.node('2')  
g.node('1')
g.node('3')
g.node('4')
g.node('8')
g.node('12')
g.node('13', fillcolor='#FFCB00', label='<<b>13</b>>', fontcolor='black', shape='doublecircle', style='filled', fixedsize='true', width='.9')

g.node('5')
g.node('9')
g.node('10')
g.node('11')
g.node('7')
g.node('6')
g.node('14')
g.node('17')
g.node('19')
g.node('24')
g.node('16')
g.node('15')
g.node('18')
g.node('20')
g.node('23')
g.node('21')
g.node('22', shape='doublecircle')


g.attr('node', shape='doublecircle', rankdir='TB', arrowsize='2.5')


g.node('22', fillcolor='#FFCB00', label='<<b>22</b>>', fontcolor='black', style='filled', fixedsize='true', width='.9')

with g.subgraph() as s:
    s.attr(rank = 'min')
    for n in ['0', '2']: s.node(n)
        
with g.subgraph() as s:
    s.attr(rank = 'same')
    for n in ['1', '3', '4', '8', '13', '12']: s.node(n)
        
with g.subgraph() as s:
    s.attr(rank = 'same')
    for n in ['9', '10', '11']: s.node(n)
        
with g.subgraph() as s:
    s.attr(rank = 'same')
    for n in ['7', '6']: s.node(n)

with g.subgraph() as s:
    s.attr(rank = 'same')
    for n in ['14', '17', '19', '24']: s.node(n)


with g.subgraph() as s:
    s.attr(rank = 'same')
    s.node('16', rank='min', shape='doublecircle', fillcolor='#FFCB00', label='<<b>16</b>>', fontcolor='black', style='filled', fixedsize='true', width='.9')

    for n in ['15', '18', '20', '23']:
        s.node(n)

with g.subgraph() as s:
    s.attr(rank = 'same')
    for n in ['21', '22']: s.node(n)

        
g.edge('0', '1')
g.edge('0', '2')
g.edge('1', '3',  minlen='2.5')
g.edge('2', '3')
g.edge('3', '4', style='dashed, bold', color='#FFCB00')
g.edge('3', '5')
g.edge('4', '6')
g.edge('7', '6')
g.edge('5', '7')
g.edge('4', '8')
g.edge('8', '9')
g.edge('9', '10')
g.edge('10', '11', minlen='2.5')
g.edge('11', '12')
g.edge('12', '13')
g.edge('6', '14')
g.edge('14', '15', style='dashed, bold', color='#FFCB00')
g.edge('16', '15') #
g.edge('14', '17')
g.edge('17', '18')
g.edge('15', '18')
g.edge('17', '19')
g.edge('10', '19')
g.edge('20', '19', style='dashed, bold', color='#FFCB00')
g.edge('20', '23',  minlen='2.5')
g.edge('11', '24')
g.edge('24', '23')
g.edge('23', '22')
g.edge('21', '22',  minlen='1.5')


g.edge('16', '7', style='invis')
g.edge('18', '20', style='invis')
g.edge('8', '13', style='invis')
g.edge('20', '17', style='invis')
g.edge('21', '20', minlen='0.5')
g.edge('19', '24', style='invis', minlen='2.5')
g.edge('8', '12', style='invis', minlen='6')

g.render(filename='img/g1')
g#.draw('graph.png')
```
